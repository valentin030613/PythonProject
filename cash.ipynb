```python
# Подключение библиотек
import pandas as pd
from bs4 import BeautifulSoup
import requests
from lxml import html

# Подключение к сайту и взятие таблицы курсов валют
url = 'https://cbr.ru/currency_base/daily/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
data = soup.find('tbody')

# Создание заголовков для таблицы с курсом валют
headers = []
for i in data.find('tr'):
    headers.append(i.text)
headers = headers[3:]
headers = headers[::2]

# Формирование строчек таблицы с курсом валют
allData = []
for i in data.find_all('tr')[1:]:
    row_data = i.find_all('td')
    row = [j.text for j in row_data]
    row = row[1:]
    allData.append(tuple(row))

# Формирование таблицы с курсом валют
myData = pd.DataFrame(allData, columns = headers)

# Вывод таблицы с курсом валют
myData.loc[:]
```
